## Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

## Load Dataset

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

### Get general info on dataset

```python
print(train.info())
```

```python
print(test.info())
```

```python
print(train.isna().sum())
```

```python
print(test.isna().sum())
```

```python
imputer = SimpleImputer(strategy = 'most_frequent')

train['fuel_type'] = imputer.fit_transform(train[['fuel_type']]).astype(str)
test['fuel_type'] = imputer.fit_transform(test[['fuel_type']]).astype(str)

train['accident'] = imputer.fit_transform(train[['accident']]).astype(str)
test['accident'] = imputer.fit_transform(test[['accident']]).astype(str)

train = train.drop('clean_title', axis = 1)
test = test.drop('clean_title', axis = 1)
```

```python
label_enc = LabelEncoder()

cat_columns = ['model', 'brand', 'fuel_type', 'engine', 'transmission', 'ext_col', 'int_col', 'accident']

for col in cat_columns:
    train[col] = label_enc.fit_transform(train[[col]].values.ravel())
    test[col] = label_enc.fit_transform(test[[col]].values.ravel())
```

```python
scaler = StandardScaler()

num_columns = ['model_year', 'milage']

for col in num_columns:
    train[col] = scaler.fit_transform(train[[col]])
    test[col] = scaler.fit_transform(test[[col]])
```

# Price Prediction

## Linear Regression Model

```python
features = ['brand', 'model', 'model_year', 'milage', 'fuel_type', 'engine', 'ext_col', 'int_col', 'accident']
target = 'price'
```

```python
lr_model = LinearRegression()

X = train[features]
y = train[target].values.ravel()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size=0.2, random_state=42)

lr_model.fit(X_train, y_train)
y_predicted = lr_model.predict(X_test)

rmse = sqrt(mean_squared_error(y_test, y_predicted))
print(f'Root Mean Squared Error: {rmse}')
```

```python
rfr_model = RandomForestRegressor(n_estimators=150, max_depth = 6, random_state=42)
rfr_model.fit(X_train, y_train)
y_predicted = rfr_model.predict(X_test)

rmse = sqrt(mean_squared_error(y_test, y_predicted))
print(f'Root Mean Squared Error: {rmse}')
```
